```python
Title = []
Artist = []
for p in soup.select('cite.title'):
     Title.append(p.get_text())
for L in soup.select('em.artist'):
     Artist.append(L.get_text())
print(Title)
print(Artist)
```

```python
# Ask the user to enter the name of a song
song_name = input("Please enter the name of a song you like: ")

# Print the song name entered by the user
print("You entered:", song_name)
```

```python
# Convert the song name to lowercase
lower_song_name = song_name.lower()
# Check if the entered song is in the DataFrame
if repertoire['Title'].str.lower().isin([lower_song_name]).any():
    recommended_songs = repertoire.loc[repertoire['Title'].str.lower() != lower_song_name, 'Title']
    print("We recommend you these songs:", recommended_songs.tolist())
else:
    print("Sorry, we don't have that song in our list.")
```
